```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries imported successfully!")
```

```python
# loding the dataset
data = pd.read_csv('loan_risk_assessment_updated.csv')
df = pd.DataFrame(data)
print("Dataset created successfully!")
print(f"Shape: {df.shape}")
```

```python
# Initial Data Exploration
print("=== DATASET OVERVIEW ===")
display(df.head())#
# Display the information about the dataset
print("\nDataset Info:")
df.info()

# Display basic statistics
print("\nBasic Statistics:")
display(df.describe())
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# DESCRIPTIVE ANALYTICS
# Average loan amount
avg_loan_amount = df['Loan_Amount'].mean()
print(avg_loan_amount)

# Average borrower age
avg_age = df['Age'].mean()
avg_age

# Income distribution
print(f"\nIncome Statistics:")
print(f"Minimum Income: ${df['Income'].min():,}")
print(f"Maximum Income: ${df['Income'].max():,}")
print(f"Average Income: ${df['Income'].mean():,.2f}")
print(f"Median Income: ${df['Income'].median():,.2f}")
print(f"Income Standard Deviation: ${df['Income'].std():,.2f}")
```

```python
# Data Observation
print("\n--- Data Observation ---")

# Most common education level
education_counts = df['Education_Level'].value_counts()
display(education_counts)
print(f"Most common: {education_counts.index[0]}")

# Most common loan purpose
print("\nLoan Purpose Distribution:")
purpose_counts = df['Loan_Purpose'].value_counts()
display(purpose_counts)
print(f"Most common: {purpose_counts.index[0]}")

# Marital status distribution
print("\nMarital Status Distribution:")
marital_counts = df['Marital_Status'].value_counts()
display(marital_counts)
print(f"Most common: {marital_counts.index[0]}")

# Credit score range
print(f"\nCredit Score Statistics:")
print(f"Range: {df['Credit_Score'].min()} - {df['Credit_Score'].max()}")
print(f"Average: {df['Credit_Score'].mean():.1f}")
print(f"Median: {df['Credit_Score'].median():.1f}")
```

```python
# Step 4: Check for missing values
print("Missing Values in Each Column:")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

# If no missing values are found, confirm:
if missing_values.sum() == 0:
    print("No missing values found in the dataset.")
else:
    print(f"Total missing values: {missing_values.sum()}")
```

```python
# checking for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Remove duplicates if any exist
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"Removed {duplicates} duplicate rows.")
    print(f"New dataset shape: {df.shape}")
```

```python
# Examine categorical variables
categorical_columns = ['Education_Level', 'Marital_Status', 'Loan_Purpose', 'Risk_Level']

print("Categorical Variables Value Counts:")
for col in categorical_columns:
    print(f"\n{col}:")
    print(df[col].value_counts())
```

```python
# Check for outliers in numerical columns using boxplots
numeric_columns = ['Age', 'Income', 'Loan_Amount', 'Interest_Rate', 'Credit_Score', 
                  'Debt_to_Income_Ratio', 'Current_Total_Debt']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
# Handle outliers using IQR method for key columns
def handle_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap the outliers
    df[column] = np.where(df[column] > upper_bound, upper_bound, 
                         np.where(df[column] < lower_bound, lower_bound, df[column]))
    return df
# Apply to columns with significant outliers
columns_to_cap = ['Income', 'Loan_Amount', 'Current_Total_Debt', 'Account_Balance']
for col in columns_to_cap:
    df = handle_outliers_iqr(df, col)
    print(f"Capped outliers in {col}")
```

```python
# Create visualizations - Distribution of Risk Levels
plt.figure(figsize=(5, 3))
risk_counts = df['Risk_Level'].value_counts()
plt.pie(risk_counts.values, labels=risk_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Loan Risk Levels')
plt.axis('equal')
plt.show()
```

```python
# Step 13: Correlation heatmap
plt.figure(figsize=(14, 10))
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Numerical Variables')
plt.tight_layout()
plt.show()
```

```python
# Distribution of Loan Purpose by Risk Level
plt.figure(figsize=(14, 8))
risk_purpose = pd.crosstab(df['Loan_Purpose'], df['Risk_Level'])
risk_purpose.plot(kind='bar', stacked=True)
plt.title('Loan Purpose Distribution by Risk Level')
plt.xlabel('Loan Purpose')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Risk Level')
plt.tight_layout()
plt.show()
```

```python
# Step 15: Age distribution by Risk Level
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Risk_Level', y='Age')
plt.title('Age Distribution by Risk Level')
plt.xlabel('Risk Level')
plt.ylabel('Age')
plt.show()
```

```python
#Employment Length distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Employment_Length', bins=20, kde=True)
plt.title('Distribution of Employment Length')
plt.xlabel('Employment Length (years)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Default History analysis
plt.figure(figsize=(10, 6))
default_risk = pd.crosstab(df['Default_History'], df['Risk_Level'])
default_risk.plot(kind='bar')
plt.title('Risk Level Distribution by Default History')
plt.xlabel('Default History (0=No, 1=Yes)')
plt.ylabel('Count')
plt.legend(title='Risk Level')
plt.show()
```

```python
#Create pairplot for key variables
key_variables = ['Age', 'Income', 'Credit_Score', 'Debt_to_Income_Ratio', 'Risk_Level']
sns.pairplot(df[key_variables], hue='Risk_Level', diag_kind='kde')
plt.suptitle('Pairplot of Key Variables by Risk Level', y=1.02)
plt.show()
```
